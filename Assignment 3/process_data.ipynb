**Preparing Data for Network**

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/umassdgithub/Fall-2023-DataViz/main/Week-8-ForceSimulator/data/data_scopus.csv")
```

```python
df[['Author(s) ID','Authors with affiliations']].iloc[0]['Authors with affiliations']
```

```python
authors = df['Author(s) ID'].apply(lambda x:x.split(";")).to_list()
```

```python
authors
```

```python
import numpy as np
```

```python
authors_flat_list = []
for items in authors:
  for author in items:
    if len(author)>0:
      authors_flat_list.append(author)
```

```python
authors_flat_list
```

```python
authors_flat_dict = {}
for author in authors_flat_list:
  authors_flat_dict[author]=df[df['Author(s) ID'].str.contains(author)]\
  [['Title','Author(s) ID','Authors','Year','Cited by','Authors with affiliations']]
```

```python
authors_flat_dict
```

```python
global errNum
errNum = 0

def getCountry(author_id):
  try:
    df_temp = authors_flat_dict[author_id]
    index = df_temp.iloc[0]['Author(s) ID'].split(";").index(author_id)
    country = df_temp.iloc[0]['Authors with affiliations'].split(";")[index].split(",")[-1].strip()
    #print(author_id,country)
    if((len(country) <= 2) or (len(country) > 15)):
      country = "United States"
    return country
  except:
    #print(df_temp)
    return -1
```

```python
#print(getCountry('36045958600'))
print(getCountry('57213664077'))
```

```python
authors_unique_list = np.unique(authors_flat_list)
len(authors_unique_list)
```

```python
countries_flat_dict = {}
for author in authors_unique_list:
  countries_flat_dict[author]=getCountry(str(author))
```

```python
#len(countries_flat_dict)
countries_flat_dict
```

***Creating Network JSON***

```python
import networkx as nx
import json
```

```python
G = nx.Graph()
```

```python
nodes = []
for id in countries_flat_dict:
  nodes.append((id,{
      "Country": countries_flat_dict[id]
  }))
```

```python
nodes
```

```python
len(nodes)
```

```python
G.add_nodes_from(nodes)
```

```python
edges = []
for id in authors_unique_list:
  current_author = id
  for row in df.iterrows():
    if ";" in row[1]['Author(s) ID']:
      authors = row[1]['Author(s) ID'][:-1].split(";")
      if authors.count(current_author) > 0:
        for author in authors:
          if author != current_author:
            edges.append((current_author,author))
  else:
    continue
```

```python
len(edges)
```

```python
G.add_edges_from(edges)
```

```python
from networkx.readwrite import json_graph
```

```python
with open("authors_network.json",'w') as f:
  json.dump(json_graph.node_link_data(G),f)
```
